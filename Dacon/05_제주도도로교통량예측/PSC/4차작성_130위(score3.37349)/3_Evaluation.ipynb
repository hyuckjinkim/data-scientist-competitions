```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
train_fn = pd.read_parquet('../OUT/train_fn.parquet.gz')
test_fn  = pd.read_parquet('../OUT/test_fn.parquet.gz')
```

```python
train_fn.head()
```

```python
train_fn['pred_ensemble'] = (train_fn['pred_lgb'] + train_fn['pred_cat']) / 2
test_fn ['pred_ensemble'] = (test_fn ['pred_lgb'] + test_fn ['pred_cat']) / 2
```

```python
from sklearn.metrics import mean_absolute_error

pred_cols = [col for col in train_fn.columns if col.find('pred_')>=0]
for col in pred_cols:
    d = train_fn.dropna()
    mae = mean_absolute_error(y_true=d['target'],y_pred=d[col])
    print(f'{col} : {mae:.4f}')
```
